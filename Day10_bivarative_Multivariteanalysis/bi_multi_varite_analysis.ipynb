```python
import pandas as pd
import seaborn as sns
```

```python
tips =sns.load_dataset('tips')
tips.head()
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
flight = sns.load_dataset('flights')
flight.head()
```

# Scatterplot Numerical -Numberical

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'], hue=tips['sex'], size=tips['size'], style =tips['smoker'])
```

# Bar Plot (Numerical - Categorical)

```python
sns.barplot(x =titanic['sex'], y =titanic['age'], hue =titanic['survived'])
```

# Box Plot (Numerical - Categorical)

```python
sns.boxplot(x=titanic['age'], y=titanic['sex'], hue=titanic['survived'])
```

# Distplot (Numerical - Categorical)

```python
sns.distplot(titanic[titanic['survived']== 0]['age'],hist=False)
sns.distplot(titanic[titanic['survived']== 1]['age'], hist=False)
```
